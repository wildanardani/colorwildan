import cv2: Mengimpor modul cv2 (OpenCV) yang digunakan untuk membaca gambar.

import matplotlib.pyplot as plt: Mengimpor modul pyplot dari Matplotlib untuk menampilkan gambar dan histogram.

imageObj = cv2.imread('gambar red green blue.jpg'): Memuat gambar dari file "gambar red green blue.jpg" ke dalam variabel imageObj menggunakan fungsi cv2.imread.

plt.axis("off"): Mengatur sumbu pada plot gambar menjadi tidak terlihat, sehingga tidak menampilkan sumbu x dan y.

plt.title("Original Image"): Memberikan judul pada plot gambar yang akan ditampilkan.

plt.imshow(cv2.cvtColor(imageObj, cv2.COLOR_BGR2RGB)): Menampilkan gambar asli dengan menggunakan fungsi imshow dari Matplotlib. Namun, karena OpenCV membaca gambar dalam format BGR, kita harus mengonversinya ke format RGB terlebih dahulu menggunakan cv2.cvtColor.

plt.show(): Menampilkan plot gambar.

blue_color = cv2.calcHist([imageObj], [0], None, [256], [0, 256]): Menghitung histogram untuk komponen warna biru (channel 0) dari gambar menggunakan fungsi cv2.calcHist. Histogram adalah representasi distribusi frekuensi intensitas piksel pada gambar.

red_color = cv2.calcHist([imageObj], [1], None, [256], [0, 256]): Menghitung histogram untuk komponen warna merah (channel 1) dari gambar.

green_color = cv2.calcHist([imageObj], [2], None, [256], [0, 256]): Menghitung histogram untuk komponen warna hijau (channel 2) dari gambar.

Berikutnya, kode tersebut akan menampilkan histogram untuk setiap komponen warna dengan menggunakan plt.hist. Di sini, kita menggunakan tiga subplot (3 baris, 1 kolom) untuk menampilkan tiga histogram secara terpisah.

plt.subplot(3, 1, 1), plt.subplot(3, 1, 2), dan plt.subplot(3, 1, 3): Mengatur posisi subplot yang akan digunakan untuk menampilkan histogram pada setiap komponen warna.

plt.title("histogram of Blue"), plt.title("histogram of Green"), dan plt.title("histogram of Red"): Menetapkan judul untuk masing-masing histogram.

plt.hist(blue_color, color="blue"), plt.hist(green_color, color="green"), dan plt.hist(red_color, color="red"): Menampilkan histogram untuk setiap komponen warna. Fungsi plt.hist digunakan untuk menampilkan histogram dengan data yang dihitung sebelumnya.

Setelah menampilkan tiga subplot histogram yang terpisah, kode ini akan menampilkan histogram gabungan untuk semua komponen warna dalam satu plot dengan menggunakan plt.hist. Namun, perlu diperhatikan bahwa ini tidak akan menghasilkan histogram yang benar, karena ketiga histogram digabung tanpa disesuaikan terlebih dahulu. Oleh karena itu, histogram gabungan ini tidak mewakili distribusi piksel yang benar dari gambar.

plt.show(): Menampilkan plot histogram gabungan.

```python

import cv2
import matplotlib.pyplot as plt

# memuat gambar
imageObj = cv2.imread('gambar red green blue.jpg')
# untuk menghindari garis line
plt.axis("off")
plt.title("Original Image")
plt.imshow(cv2.cvtColor(imageObj, cv2.COLOR_BGR2RGB))
plt.show()

# mendapatkan data rgb dari gambar
blue_color = cv2.calcHist([imageObj], [0], None, [256], [0, 256])
red_color = cv2.calcHist([imageObj], [1], None, [256], [0, 256])
green_color = cv2.calcHist([imageObj], [2], None, [256], [0, 256])

# memisahkan histogram pada setiap warna
plt.subplot(3, 1, 1)
plt.title("histogram of Blue")
plt.hist(blue_color, color="blue")

plt.subplot(3, 1, 2)
plt.title("histogram of Green")
plt.hist(green_color, color="green")

plt.subplot(3, 1, 3)
plt.title("histogram of Red")
plt.hist(red_color, color="red")

# menampilkan tampilan yang jelas
plt.tight_layout()
plt.show()

# menyatukan histogram setiap warna
plt.title("Histogram of all RGB Colors")
plt.hist(blue_color, color="blue")
plt.hist(green_color, color="green")
plt.hist(red_color, color="red")
plt.show()
```
